```python
import pandas as pd
import numpy as np
import time
```

### Arauz uni√≥n de datos

```python
df_last_arauz = pd.read_csv("Twitter_search_Arauz.csv") 
#df_last_arauz.head(5)
df_last_arauz['created_at'] = pd.to_datetime(df_last_arauz['created_at'], format="%a %b %d %H:%M:%S +0000 %Y")
df_last_arauz = df_last_arauz[df_last_arauz['created_at']<=time.strftime('%Y-%m-%d %H:%M:%S', time.strptime('Mon Feb 01 00:00:00 +0000 2021','%a %b %d %H:%M:%S +0000 %Y'))]
df_last_arauz.head(5)
```

```python
df_new_arauz = pd.read_csv("Twitter_search_Arauz_new.csv") 
df_new_arauz['created_at'] = pd.to_datetime(df_new_arauz['created_at'], format="%a %b %d %H:%M:%S +0000 %Y")
df_new_arauz = df_new_arauz[df_new_arauz['created_at']>time.strftime('%Y-%m-%d %H:%M:%S', time.strptime('Mon Feb 01 00:00:00 +0000 2021','%a %b %d %H:%M:%S +0000 %Y'))]
df_new_arauz = df_new_arauz[df_new_arauz['created_at']<time.strftime('%Y-%m-%d %H:%M:%S', time.strptime('Sun Feb 08 00:00:00 +0000 2021','%a %b %d %H:%M:%S +0000 %Y'))]
df_new_arauz.head(5)
```

```python
data_tweets_arauz = pd.concat([df_last_arauz, df_new_arauz], axis=0)
data_tweets_arauz = data_tweets_arauz.reset_index(drop=True)
data_tweets_arauz.head(5)
```

```python
data_tweets_arauz.to_csv("all_tweets_arauz.csv",index=False)
```

```python
len(data_tweets_arauz)
```
