## **2.1.&emsp; Buenas prácticas**
>Pandas es muy eficiente con datos pequeños (generalmente de 100 MB a 1 GB) y el rendimiento rara vez es una preocupación. De hecho, Pandas tiene su propia limitación cuando se trata de big data debido a su algoritmo y restricciones de memoria local. Por lo tanto, los grandes datos generalmente se almacenan en clústeres informáticos para una mayor escalabilidad y tolerancia a fallas. Y a menudo se puede acceder a través del ecosistema de big data (AWS EC2, Hadoop, etc.) usando Spark y muchas otras herramientas. **Eventualmente, una de las formas de usar Pandas con datos grandes en máquinas locales (con ciertas restricciones de memoria) es reducir el uso de memoria de los datos.**</br>
Para el manejo de grandes datos se han ideado [buenas prácticas](https://towardsdatascience.com/why-and-how-to-use-pandas-with-large-data-9594dda2ea4c) para mejorar el rendimiento de Pandas, las cuales predicamos en este documento:
- **Leer datos del archivo CSV con Chunksizes**</br>
&emsp;&emsp;Significa el número de filas que se leerán en un marco de datos en cualquier momento para que quepan en la memoria local.
- **Filtrar columnas sin importancia para ahorrar memoria**</br>
&emsp;&emsp;Para ahorrar más tiempo para la manipulación y el cómputo de datos, se filtran algunas columnas sin importancia.
- **Cambiar el tipo de dato de las columnas**</br>
&emsp;&emsp;Al reducir los bits necesarios para almacenar los datos, se reduce el uso general de la memoria hasta en un 50%
- **Almacenamiento en disco I/O**</br>
&emsp;&emsp;Para mejorar el rendimiento de los procesos I/O en los dataframes utilizados. Se pueden usar [formatos binarios](https://towardsdatascience.com/the-best-format-to-save-pandas-data-414dca023e0d) para almacenar los datos &emsp;&emsp;en el disco, y muchos de ellos son compatibles con pandas. En este proyecto utilizaremos el formato .pkl (Pickle)

## **4.2.&emsp;Preparar el modelo**
#### **4.2.1.&emsp;Correlation Treshold**
>Los umbrales de correlación eliminan características que están altamente correlacionadas con otras (es decir, sus valores cambian de manera muy similar a la de los demás). Estas características proporcionan información redundante. Primero calculamos todas las correlaciones por pares. Luego, si la correlación entre un par de características está por encima de un umbral dado, eliminaremos la que tenga una correlación media absoluta más grande con otras características.

```python
#----------------------- Logistic Regression Model ------------------#
from sklearn import linear_model                                     #
from sklearn import model_selection                                  #
from sklearn.model_selection import train_test_split                 #
from sklearn.feature_selection import RFE                            #
from sklearn.preprocessing import MinMaxScaler                       #
from sklearn.metrics import classification_report                    #
from sklearn.metrics import confusion_matrix                         #
from sklearn.metrics import accuracy_score                           #
#--------------------------------------------------------------------#
```

```python
X = df_students.drop(['G3','class'],axis=1)
y = df_students['class']

scaler = MinMaxScaler()
tmp = scaler.fit_transform(X)
X = pd.DataFrame(tmp, columns=X.columns)

X_train, X_test, y_train, y_test = train_test_split( X, y,
                                                 test_size=.25,
                                                 stratify=y,
                                                 random_state=1)

X_train.shape,X_test.shape
```

```python
col_corr = set()
# Finding the correlation between columns.
corr_matrix = X.corr()

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        # Higher correlation
        if abs(corr_matrix.iloc[i,j]) > 0.6:
            # getting the column name
            colName = corr_matrix.columns[i]
            # adding the correlated column name higher than treshold value (0.8)
            col_corr.add(colName)
            
print('Correlated columns: ', col_corr)
```

#### **4.2.2.&emsp;Feature Selection - Recursive Feature Elimination**
>El método  Recursive Feature Elimination funciona eliminando recursivamente los atributos y construyendo un modelo sobre los atributos que quedan. Utiliza métrica de precisión para clasificar la característica según su importancia. El método RFE toma el modelo a utilizar y la cantidad de características requeridas como entrada. Luego da la clasificación de todas las variables, siendo 1 la más importante. También brinda su soporte, True es una característica relevante y False es una característica irrelevante.

```python
model = linear_model.LogisticRegression(solver='lbfgs', multi_class='auto',
                                       max_iter=5000)

cols = list(X.columns)
# Se escogen 13 posible variables, este valor puede cambiar
rfe = RFE(model, 15)
#Transforming data using RFE
X_rfe = rfe.fit_transform(X,y)  
#Fitting the data to model
model.fit(X,y)
print(rfe.support_)
print(rfe.ranking_)
temp = pd.Series(rfe.support_,index = cols)
selected_features_rfe = temp[temp==True].index
print(selected_features_rfe)
```

#### **4.2.3.&emsp;Feature Selection - Forward Feature Selection**
>En la primera fase de la selección de la función de avance, se evalúa el rendimiento del clasificador con respecto a cada función. La función que rinde mejor se selecciona de todas las funciones.
En el segundo paso, la primera característica se prueba en combinación con todas las demás características. Se selecciona la combinación de dos características que producen el mejor rendimiento del algoritmo. El proceso continúa hasta que se selecciona el número especificado de características.

```python
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
# Use RandomForestRegressor algoritham as an estimator. 
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
```

```python
model = sfs(RandomForestRegressor(n_estimators=100), 
            k_features=15, 
            forward=True,
            verbose=2, 
            cv=5, 
            n_jobs=-1, 
            scoring='r2')

model.fit(X_train, y_train)
```

```python
best_features = list(model.k_feature_names_)
best_features
```

#### **4.2.4.&emsp;Feature Selection - Backward Feature Selection**
>Como su nombre lo indica, al principio le brindamos todas las características posibles al modelo. Verificamos el rendimiento del modelo y luego eliminamos iterativamente las características de peor rendimiento una por una hasta que el rendimiento general del modelo se encuentre en un rango aceptable.
La métrica de rendimiento utilizada aquí para evaluar el rendimiento de la característica es pvalue. Si el valor p está por encima de 0.05, eliminamos la característica, de lo contrario la conservamos.

```python
import statsmodels.api as sm

#Backward Elimination
cols = list(X.columns)
pmax = 1
while (len(cols)>0):
    p= []
    X_1 = X[cols]
    X_1 = sm.add_constant(X_1)
    model = sm.OLS(y,X_1).fit()
    p = pd.Series(model.pvalues.values[1:],index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
selected_features_BE = cols
print(selected_features_BE)
```
